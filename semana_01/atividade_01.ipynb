```python
# Faz uma busca a partir de um novo índice da linha, então retorna todos os dados a partir do valor declarado
# Valor antes da vírgula é o índice da linha, após a vírgula é o índice da coluna
fb.iloc[3:, : ]
```

```python
# O loc está fazendo um slice dos dados por labels
plt.figure(figsize=(12, 8))
fb.loc['2015-01-01': '2015-12-31', 'Close'].plot()
fb.loc['2016-01-01': '2016-12-31', 'Close'].plot()
fb.loc['2017-01-01': '2017-12-31', 'Close'].plot()
fb.loc['2018-01-01': '2018-12-31', 'Close'].plot()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
fb['Open'].plot()
plt.show()
```
